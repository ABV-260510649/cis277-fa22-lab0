## While Loop (less Pythonic)

```python
s = "RacecAr "

def is_palindrome1(string):
    clean_string = string.lower().replace(' ', '')
    first, last = 0, len(clean_string)-1
    while( first < last):
        if clean_string[first] == clean_string[last]:
            first +=1
            last -=1
        else:
            return False
    
    return True

is_palindrome1(s) 
```

## String Slicing (More Pythonic)

```python
def is_palindrome2(string):
    clean_string = string.lower().replace(' ', '')
    reversed_string = clean_string[::-1]
    return clean_string == reversed_string

s = " Rac eC ar "
is_palindrome2(s)
```

```python
import csv

ratings = {}
directors = {}
plots = {}

with open('movie.csv', encoding="utf-8") as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        #print(row['movie_title'], row['imdb_score'])
        ratings[row['movie_title']] = float(row['imdb_score'])
        directors[row['director_name']] = 0
        plots[row['genres']] = row['duration']

```

What is the average IMDB Score of all the movies?

```python
sum = 0

for rating in ratings.values():
    sum += rating

average_rating = sum / len(ratings)
print(average_rating)

```

### Questions
1. How many movies have "Star Wars" in the movie_title field?

```python
star_wars_amount = 0
for titles in ratings.keys():
    if "Star Wars" in titles:
        star_wars_amount += 1
print(star_wars_amount)
```

2. Which movie has the highest IMDB Score?

```python

highest_score = max(ratings.values())
for key,value in ratings.items():
    if highest_score == value:
        print(key)
```

3. Which movie has the lowest IMDB Score?


```python
lowest_score = min(ratings.values())
for key, value in ratings.items():
    if lowest_score == value:
        print(key)
```

5. What's the average duration of Action movies?

```python
action_movies = []
duration_length = []
total_length = 0

for movie_types in plots.keys():
    if "Action" in movie_types:
        action_movies.append(movie_types)

for key, value in plots.items():
    if key in action_movies:
        if value != '':
            duration_length.append(float(value))

for number in duration_length:
    total_length += number

print(total_length / len(duration_length))



```
